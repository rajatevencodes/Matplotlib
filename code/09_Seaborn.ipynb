#### **Data Visualization with Seaborn** âœ¨

**Seaborn** is a high-level Python visualization library that is built directly on top of **Matplotlib**. Its main purpose is to provide a simpler interface for creating beautiful and statistically informative graphics.

While Matplotlib is powerful, it can be very low-level and often requires significant code for customization. Seaborn simplifies this with several key advantages.

#### **Why Use Seaborn?**

* **Beautiful Aesthetics by Default:** Seaborn comes with professional-looking themes and smart color palettes built-in, so your charts look great with minimal effort.
* **Seamless DataFrame Integration:** It is designed to work perfectly with Pandas `DataFrames`, allowing you to create complex plots directly from your data with simple syntax.
* **Simplified Complex Plots:** Creating advanced statistical plots like **boxplots**, **violin plots**, **heatmaps**, and **pairplots** can be done in just a single line of code.
* **Built-in Datasets:** It includes several sample datasets, making it easy to practice and learn.

In short, Seaborn allows you to write **less code** to produce **better-looking graphs**, especially when working with Pandas `DataFrames`.

```python
import seaborn as sb

# Seaborn comes with several built-in datasets.
# See all available datasets
# print(sb.get_dataset_names())

# We can load them using the `load_dataset` function.
tips = sb.load_dataset("tips")
# print(tips.head()) # Check the data
sb.set_theme(style="darkgrid") 

import matplotlib.pyplot as plt

# Line Plot

sb.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
# Scatter Plot
sb.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# Bar Plot
sb.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
# Box Plot
sb.boxplot(x="day", y="total_bill", data=tips)
plt.title('Boxplot of Total Bill per Day')
plt.show()
```

```python
## Heatmap
corr=tips[['total_bill','tip','size']].corr()
sb.heatmap(corr,annot=True,cmap='coolwarm')
```
