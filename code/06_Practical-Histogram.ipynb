| Parameter | Purpose |
|----------|---------|
| `bins` | Number or custom edges of bins |
| `edgecolor` | Color around each bar | 
| `axvline` | Vertical reference line | 

A histogram is the perfect tool when you want to understand the distribution of a single numerical variable, like the ages of your YouTube viewers. Instead of looking at thousands of individual ages, a histogram groups them into "bins" (e.g., ages 13-18, 18-25, etc.) and shows you how many viewers fall into each age bracket.

This allows you to instantly see:
- The most common age group of your audience.
- The spread of ages (are they all young, or is there a wide range?).
- Any skewness (e.g., are your viewers disproportionately young?).


```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42) 
viewer_ages = np.random.randint(13, 60, size=500)

# bins or bars :)
age_bins = [10, 18, 25, 35, 45, 55, 65]

plt.figure(figsize=(10, 6))

# The main histogram function 
plt.hist(viewer_ages, bins=age_bins, edgecolor='black')

# Add a Vertical Line for the Average Age
average_age = np.mean(viewer_ages)
# This function is not part of the histogram itself but is used to draw a vertical line on your plot. It's excellent for marking key values like the mean, median, or a specific target.
plt.axvline(average_age, color='red', linestyle='--', linewidth=2, label=f'Average Age: {average_age:.1f}')

# Add Labels and a Title for Clarity
plt.title('Distribution of YouTube Viewer Ages')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.legend()

# Show the Final Plot
plt.show()
```
